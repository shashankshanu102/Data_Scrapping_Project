```python
def get_data(url):
    
    from selenium import webdriver
    import os
    from bs4 import BeautifulSoup
    
    path_tounzipfile = "C:/Users/SHASHANK/Desktop/DATA SCRAPP"
    os.chdir(path_tounzipfile)
    browser = webdriver.Firefox(path_tounzipfile)
    
  
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html,"html.parser")
    
    image_url = []
    for i in soup.find_all("img"):
        image_temp = i.get('src')
        if '.jpg' in image_temp:
            image_url.append(image_temp)
        
    name = []
    for i in soup.find_all("h3", class_="lister-item-header"):
        name.append(i.find_all("a")[0].text)
        name = list(map(lambda s: s.strip(), name))
        
    trait = []
    all_d = soup.find_all("div",class_="lister-item-content")
    for i in all_d:
        paras = i.find_all("p")
        trait.append(paras[1].text.strip())
    
    
    df = pd.DataFrame({"NAME":name,"IMAGE_URL":image_url,"TRAIT":trait})
    
    return df
```

```python
# calling the user defined fuction and returning the resultant dataframe with necessary informations scrapped from the web page.
url = "https://www.imdb.com/list/ls025929404/"
get_data(url)
```

### Selenium advantages :
- Using selenium we can depict a full fledged web browser.
- This will help us to extract data from many pages of the same website.
- So using various fuctions of selenium we can hover through all the pages and extract data from it.
- Thus using selenium we can click on the buttons on a particular website to perform certain action from remote user.
- In the below code we are finding elements by class name which consists of button, and the clicking on the button using .click()
- In this way we can build a full fledged web browser and control it form remote user using selenium.

### Creating Final Dataframe :
- This dataframe will contain all the informations scrapped using the above user defined function.
- In the below code we are first creating an empty dictionary with key and value pair.
- Keys will be the page numbers and values will be the data frame corresponding to the particular page.
- We are using the .click() function to hover through the pages and extract data.
- so the resultant dictionary will consist of many dataframes with desired information scrapped from all the pages.

```python
url2 = "https://www.imdb.com/list/ls025929404/?sort=list_order,asc&mode=detail&page={}"
browser.get(url2)

main={}

for i in range(1,3):
    main[i]=get_data(url2.format(i))
    
    btns2=browser.find_elements_by_class_name("flat-button lister-page-next next-page")
    for i in btns2:
        if i== "NEXT":
             btns2.click()
```

```python
# Here we are concating all the dataframes present in the above formed dictionary for further analysis. 

import pandas as pd
big_data1= pd.concat([main[i] for i in main],axis =0)
```

```python
# Reseting the index of the above formed final data frame consisting of information stored in it extracted from all the pages.

big_data1=big_data1.reset_index()
big_data1.drop("index",inplace=True,axis=1)
```

```python
def get_data2(url):
    
    from selenium import webdriver
    import os
    from bs4 import BeautifulSoup
    
    path_tounzipfile = "C:/Users/SHASHANK/Desktop/DATA SCRAPP"
    os.chdir(path_tounzipfile)
    browser = webdriver.Firefox(path_tounzipfile)
    
  
    browser.get(url)
    html = browser.page_source
    soup1 = BeautifulSoup(html,"html.parser")
    
    img_url=[]
    for i in soup1.find_all("figure",class_="listItem__figure relative"):
        for i in soup1.find_all("img"):
            dt_src = i.get("data-src")
            #dt_src = list(dt_src)
            #print(dt_src)
            img_url.append(dt_src)
    only_url=[]
    for i in img_url:
        if i == None:
            continue
        else:
            only_url.append(i)
    
    ## Extracting names of the celeberities
    name=[]
    for i in soup1.find_all("div",class_="listItem"):
        for j in soup1.find_all("a",class_="listItem__title"):
            name.append(j.text)
        

        
    ## Extracting description of the celeberities

    trait=[]
    for i in soup1.find_all("div",class_="listItem"):
        for j in soup1.find_all("span",class_="listItem__wiki block default grey"):
            trait.append(j.text.strip())
    
    df2=pd.DataFrame({"Name":pd.Series(name),"IMAGE_URL":pd.Series(only_url),"TRAIT":pd.Series(trait)})
    
    
    return df2
```

- When we concatenate DataFrames, we need to specify the axis. axis=0 tells pandas to stack the second DataFrame UNDER the first one. It will automatically detect whether the column names are the same and will stack accordingly. axis=1 will stack the columns in the second DataFrame to the RIGHT of the first DataFrame. To stack the data vertically, we need to make sure we have the same columns and associated column format in both datasets. When we stack horizontally, we want to make sure what we are doing makes sense (i.e. the data are related in some way).
